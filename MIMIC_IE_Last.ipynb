```python
# Removing the outliers
def removeOutliers(f):
    filt_df = f.loc[:, f.columns != 'HADM_ID']
    low = .1
    high = .9
    quant_df = filt_df.quantile([low, high])
    
    print(quant_df)
     
    filt_df = filt_df.apply(lambda x: x[(x>quant_df.loc[low,x.name]) &
                                    (x < quant_df.loc[high,x.name])], axis=0)
        
    filt_df = pd.concat([f.loc[:,'HADM_ID'], filt_df], axis=1)
    print(filt_df.shape)
    return filt_df
```

```python
ck = []
for k in v:
    f = grouped.get_group(k)
    print("HADM_ID %s" %k)
    ck.append(removeOutliers(f))
     
```

```python
ot=pd.concat(ck)
ot.shape
```

```python
LIST = ['DEXTROSE 5%',  
         'NACL 0.9%', 'NOREPINEPHRINE', 'PO INTAKE']
```

```python
for i in LIST:
    ot[i] = ot[i].fillna(ot.groupby(['HADM_ID'])[i].transform('mean'))
    ot[i] = ot.groupby(['HADM_ID'])[i].transform('mean')
```
