```python
import datetime
#time convertion
Data1['DOB']=pd.to_datetime(Data1['DOB'],errors='coerce')
Data1['DATE']=pd.to_datetime(Data1['DATE'],errors='coerce')
```

```python
Data1['age'] = round((Data1['DATE']- Data1['DOB']).apply(lambda x : x.days/365.242),2)
Data1['age'].value_counts()
```

```python
def neg_conv(s):
    if s < 0:
        s = s+300
    return s
Data1['age'] = Data1['age'].apply(lambda x: neg_conv(x))
Data1['age']
```

```python
def mapping_to_bin(s):
    if s <= 14:
        s = 'Child'
    elif 14 < s <= 89:
        s = 'Adult'
    return s
Data1['age_group'] = Data1['age'].apply(lambda x: mapping_to_bin(x))
Data1['age_group']
```
