```python
# Removing the outliers
def removeOutliers(f):
    filt_df = f.loc[:, f.columns != 'HADM_ID']
    low = 0.1
    high = 0.9
    quant_df = filt_df.quantile([low, high])
    
    print(quant_df)
    print(filt_df.shape) 
    print(filt_df.isna().sum())
    filt_df = filt_df.apply(lambda x: x[(x>quant_df.loc[low,x.name]) & 
                                    (x < quant_df.loc[high,x.name])], axis=0)
    print(filt_df.shape) 
    print(filt_df.isna().sum())
    filt_df = pd.concat([f.loc[:,'HADM_ID'], filt_df], axis=1)
    print(filt_df.isna().sum())
    print(filt_df.shape)
    return filt_df
```
